```python
from matplotlib import pylab
print(pylab.__version__)
```

```python
import numpy as np

x = np.linspace(0, 10, 25)
y = x*x+2
print(x)
print(y)
```

```python
print(np.array([x, y]).reshape(25, 2))
```

```python
pylab.plot(x, y, 'r')
```

```python
pylab.plot(x, y, 'b')
```

```python
pylab.subplot(1,1,1)
pylab.plot(x, y, 'r--')
pylab.subplot(1,2,1)
pylab.plot(y, x,'g*--')
```

```python
pylab.subplot(1,1,1)
pylab.plot(x, y, 'r--')
pylab.subplot(2,2,1)
pylab.plot(y, x,'g*--')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x, y, 'r--')
pylab.subplot(2,2,2)
pylab.plot(y, x,'g*--')
```

```python
pylab.subplot(1,2,1)
pylab.plot(x, y, 'r--')
pylab.subplot(2,2,2)
pylab.plot(y, x,'g*--')
pylab.subplot(3,2,3)
pylab.plot(y, x,'g*--')
```
